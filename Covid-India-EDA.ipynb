```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
data
```

# Eliminating the last row since we don't need it and first column because we will use dataframe index as Serial number

```python
data = data.iloc[:-1,1:] #this eliminates the last row and first column from the dataframe.
```

```python
data.tail()
```

```python
data.head()
```

```python
total_state = np.arange(len(data['state_name']))
# data['state_name']
```

```python
total_state
```

# Positive cases VS states in India

```python
from matplotlib.pyplot import figure
sns.set()
figure(num=None, figsize=(16,6), dpi=150, facecolor= (0.8,0.8,0.9), edgecolor='b')
plt.bar(data['state_name'],data['positive'], align='center', alpha=0.8, color=(0.9,0.2,0.2), edgecolor=(0,0,0))
plt.xticks(total_state, data['state_name'], rotation='vertical')
plt.ylim(1,max(data['positive'])+100000)
plt.title('Positive Cases')
# plt.annotate()
plt.show()
plt.tight_layout()
```

# Active cases VS states in India

```python
from matplotlib.pyplot import figure
sns.set()
figure(num=None, figsize=(16,6), dpi=150, facecolor= (0.8,0.8,0.9), edgecolor='b')
plt.bar(data['state_name'],data['active'], align='center', alpha=0.8, color=(0.5,0.8,0.2), edgecolor=(0,0,0))
plt.xticks(total_state, data['state_name'], rotation='vertical')
plt.ylim(1,max(data['active'])+100000)
plt.title('Active Cases')
# plt.annotate()
plt.show()
plt.tight_layout()
```
